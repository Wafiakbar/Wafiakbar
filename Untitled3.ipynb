```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(style="whitegrid")

# Data
years = [2024, 2025]
world_growth = [3.2, 2.9]
us_growth = [2.1, 1.6]
china_growth = [4.8, 4.2]
indonesia_growth_range = (4.7, 5.5)
indonesia_growth_est = 4.9

exports_data = {
    "Negara": ["AS", "Tiongkok", "ASEAN", "Jepang"],
    "Ekspor (Miliar USD)": [18.5, 27.3, 36.2, 14.1]
}
exports_df = pd.DataFrame(exports_data)

regions = ["Jawa", "Kalimantan", "Sumatera", "Sulawesi", "Papua"]
growth_rate = [5.1, 5.3, 4.8, 4.6, 3.9]

# Data tambahan
df = pd.DataFrame({
    'Tahun': [2024, 2025],
    'Dunia': [3.2, 2.9],
    'Indonesia': [5.0, 4.9]
})
df_melt = df.melt(id_vars='Tahun', var_name='Kategori', value_name='Pertumbuhan')

# Data konsumsi rumah tangga
data_konsumsi = {
    'Tahun': [2024, 2025],
    'Konsumsi_Rumah_Tangga': [4.8, 4.6]
}
df_konsumsi = pd.DataFrame(data_konsumsi)

# Data PDB
sektor = ['Industri', 'Pertanian', 'Jasa', 'Pertambangan']
kontribusi = [40, 15, 35, 10]

# Membuat satu figure besar
fig, axs = plt.subplots(3, 2, figsize=(18, 15))

# 1. Scatter Plot - Pertumbuhan Ekonomi Global
axs[0, 0].scatter(years, world_growth, label="Dunia", color='darkorange', s=100)
axs[0, 0].scatter(years, us_growth, label="AS", color='orange', s=100)
axs[0, 0].scatter(years, china_growth, label="Tiongkok", color='orangered', s=100)
axs[0, 0].set_title("Pertumbuhan Ekonomi Global (%) - Scatter Plot")
axs[0, 0].set_xlabel("Tahun")
axs[0, 0].set_ylabel("Persentase")
axs[0, 0].legend()

# 2. Horizontal Bar Chart - Ekspor Nonmigas
sns.barplot(y="Negara", x="Ekspor (Miliar USD)", data=exports_df, ax=axs[0, 1], palette="Oranges_r")
axs[0, 1].set_title("Ekspor Nonmigas Indonesia Triwulan I 2025 (Horizontal)")

# 3. Stacked Bar Chart - Pertumbuhan Wilayah
stack_df = pd.DataFrame({'Wilayah': regions, 'Pertumbuhan': growth_rate})
axs[1, 0].barh(stack_df['Wilayah'], stack_df['Pertumbuhan'], color='darkorange')
axs[1, 0].set_title("Pertumbuhan Ekonomi Wilayah (Stacked Style)")

# 4. Bullet Chart Style - Proyeksi Indonesia
axs[1, 1].barh(["Proyeksi BI 2025"], [5.5], color='navajowhite', height=0.6)
axs[1, 1].barh(["Proyeksi BI 2025"], [indonesia_growth_est], color='darkorange', height=0.3)
axs[1, 1].set_xlim(4.5, 5.7)
axs[1, 1].set_title("Bullet Style: Proyeksi Pertumbuhan Ekonomi Indonesia 2025")

# 5. Clustered Scatter Plot - Perbandingan Ekonomi
for tahun, style in zip([2024, 2025], ['o', 's']):
    subset = df_melt[df_melt['Tahun'] == tahun]
    axs[2, 0].scatter(subset['Kategori'], subset['Pertumbuhan'], label=f"Tahun {tahun}", s=200, marker=style)
axs[2, 0].set_title('Perbandingan Pertumbuhan Ekonomi Global dan Indonesia - Clustered Scatter')
axs[2, 0].legend()

# 6. Donut Chart - Kontribusi Sektor
wedges, texts, autotexts = axs[2, 1].pie(kontribusi, labels=sektor, autopct='%1.1f%%',
                                         startangle=140, colors=['orange', 'darkorange', 'gold', 'peru'],
                                         wedgeprops=dict(width=0.4))
axs[2, 1].set_title('Kontribusi Sektor ke PDB Indonesia 2025 (Donut)')

plt.tight_layout()
plt.show()

# Line Chart Konsumsi Rumah Tangga
plt.figure(figsize=(8,6))
plt.plot(df_konsumsi['Tahun'], df_konsumsi['Konsumsi_Rumah_Tangga'], marker='o', linestyle='--', color='darkorange')
plt.title('Pertumbuhan Konsumsi Rumah Tangga Indonesia 2024-2025 (Line Plot)')
plt.ylabel('Persentase (%)')
plt.xlabel('Tahun')
plt.grid(True)
plt.show()
```
